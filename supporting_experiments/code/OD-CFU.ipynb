# OD-CFU correlation measurements data analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import matplotlib
import matplotlib.ticker
from PARAMS import *
%run -i PARAMS.py
```

```python
mf = matplotlib.ticker.ScalarFormatter(useMathText=True)
mf.set_powerlimits((-2,2))
```

```python
data = pd.read_csv('../data/2022-12-08_OD-CFU_16.csv', sep=',')
data['mean'] = data[['16-1','16-2','16-3']].mean(axis=1)
data['err'] = data[['16-1','16-2','16-3']].std(axis=1)
data
```

```python
i = 0
for cond in ['centrifuged','not-centrifuged']:
    new_dat = data[['focus', 'cond', 'mean', 'err']][data['cond']==cond].set_index('focus').drop('cond', axis=1).T
    plt.errorbar(new_dat['OD']['mean'],new_dat['CFU']['mean'], color=cb[i], yerr=new_dat['CFU']['err'], xerr=new_dat['OD']['err'],fmt='o',label=cond)
    i += 1
plt.legend()
plt.xticks(np.arange(0.2,0.62, 0.1))
plt.ylim(0,7*10**7)
plt.ylabel('CFU')
plt.xlabel('OD')
plt.gca().yaxis.set_major_formatter(mf)
plt.tight_layout()
plt.savefig('../figures/2022-12-08_OD-CFU_16.png')
```

## Visualization in one plot

Add all data to one dataframe and plot as one plot to save sapce in the thesis.

```python
data = pd.read_csv('../data/2022-12-21_OD-CFU_14-16.csv', sep=',')
data.replace({'cent log': 'centrifuged', 'not cent log': 'not-centrifuged'}, inplace=True)
data = data[data['cond'].isin(['centrifuged', 'not-centrifuged'])]
data.reset_index(drop=True, inplace=True)
```

```python
data2 = pd.read_csv('../data/2023-01-11-OD-CFU_15-17.csv', sep=',')
```

```python
complete = data.join(data2.drop(['focus', 'cond'], axis=1))
complete = complete.melt(id_vars=['focus', 'cond'])
complete[['strain', 'replicate']] = complete.variable.str.split('-', expand=True)
```

```python
foc = [v for k, v in complete.groupby('focus')]
df = foc[0].rename({'value': 'CFU'}, axis=1).drop(['focus', 'variable'], axis=1)
od = foc[1].rename({'value': 'OD'}, axis=1).drop(['focus', 'variable'], axis=1)
all = pd.merge(df, od, on=['cond', 'strain', 'replicate'])
all.rename({'cond': 'condition'}, axis=1, inplace=True)
all['strain'].replace(STRAINS_LAB, inplace=True)
all
```

```python
g = sns.scatterplot(
    all,
    x = 'OD',
    y= 'CFU',
    hue = 'strain',
    style='condition',
    hue_order=STRAINS_LAB.values(),
    s=100, 
    palette=cs2
)
g.legend(loc='center left', bbox_to_anchor=(1.0, 0.5), ncol=1)
plt.savefig('../figures/OD-CFU_all.png', bbox_inches="tight")
```
